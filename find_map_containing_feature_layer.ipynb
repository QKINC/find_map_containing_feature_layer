```python
# Imports
from arcgis.gis import GIS
from arcgis.mapping import WebMap
from getpass import getpass
```

```python
# Run this to Login
try:
    portal = input("What is the portal name?")
    username = input("what is your username?")
    print("Please enter your password...")
    password = getpass()


    parameters = {
        "PortalUrl" : f"https://{portal}.maps.arcgis.com/",
        "PortalUserName" : f"{username}",
        "PortalPassword" : f"{password}"
    }

    for keys, values in parameters.items():
        url = parameters['PortalUrl']
        username = parameters['PortalUserName']
        password = parameters['PortalPassword']

    gis = GIS(url, username, password)

    print("Successfully logged into")
except Exception as e:
    print(f"Could not log in: {e}")
    

# token = gis._con.token
# print(f"Token is: {token}")ddd
```

```python
# Create the method
def find_maps_with_feature_layer(feature_item_id):
    # Get the feature layer by ID
    feature_layer = gis.content.get(feature_item_id)

    # Get all the web maps in the portal
    items = gis.content.search(query="", item_type="Web Map", max_items = 5000)
    total = len(items)
    
    print(f"Searching through {total} items...")
    
    # List to store maps containing the feature layer
    maps_with_feature_layer = []

    # Iterate through each item
    x = 0
    for item in items:
        try:
            print(f"Searching item {item.title}: {x}/{total}...")
            # Create a WebMap object
            web_map = WebMap(item)
            # Iterate through each layer in the WebMap
            x+=1
            for layer in web_map.layers:
                # Check if the layer is the feature layer
                if 'url' in layer and feature_layer.url in layer['url']:
                    # Append the web map to the list
                    maps_with_feature_layer.append(item)
                    print("FOUND A MAP WITH ITEM APPENDING TO LIST...\n")
                    break
        except Exception as e:
            print(f"Failed to run; {e}")
    
    # Return the list of maps containing the feature layer
    return maps_with_feature_layer
```

```python
item_id = input("What is the item ID for the feature you would like to find?")
maps = find_maps_with_feature_layer(f'{item_id}')
print("The maps that contain this feature layer are as follows...")
print(maps)
```
